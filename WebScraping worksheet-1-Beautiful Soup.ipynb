### 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’¶


```python
def Wikipedia(url):
    Header=[]
    wikipage=requests.get(url)
    wiki_soup = BeautifulSoup(wikipage.content)
    
    header = wiki_soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
    for i in header:
        Header.append(i.text.replace("\n",""))
    Wiki={'Header Tag':Header}
    Wiki=pd.DataFrame(data=Wiki)
    return(Wiki)
```

```python
Wikipedia('https://en.wikipedia.org/wiki/Main_Page')
```

### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.


```python
imdbpage=requests.get('https://www.imdb.com/chart/top/')
imdbpage
```

```python
imdbsoup=BeautifulSoup(imdbpage.content)
imdbname=imdbsoup.find('td',class_='titleColumn')
movie_title=imdbname.find('a').get_text()
movie_title
```

```python
imdbrating=imdbsoup.find('td',class_="ratingColumn imdbRating")
imdbrating.text.replace('\n','')
```

```python
imdbyear=imdbsoup.find('span',class_="secondaryInfo")
imdbyear.text.replace('(','').replace(')','')
```

```python

def IMDB_top_100(url):
    Name=[]
    IMDB_rating=[]
    Released_year=[]
    imdbsoup=BeautifulSoup(imdbpage.content)
    imdbname=imdbsoup.find_all('td',class_='titleColumn')
    imdbrating=imdbsoup.find_all('td',class_="ratingColumn imdbRating")
    imdbyear=imdbsoup.find_all('span',class_="secondaryInfo")
    
    for i in imdbname:
        movie_title=i.find('a').get_text()
        Name.append(movie_title)
    for i in imdbrating:
        IMDB_rating.append(i.text.replace('\n',''))
    for i in imdbyear:
        Released_year.append(i.text.replace('(','').replace(')',''))
        
    Top_100=pd.DataFrame({})
    Top_100['Names']=Name
    Top_100['IMDB Rating']=IMDB_rating
    Top_100['Year of release']=Released_year
    Top_100=Top_100[:100]
    return(Top_100)
```

### 3.Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
def getIndianTopMovie(url,element,class1):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    movieName=soup.find_all(element,class_=class1)
    IndianName_list=[]
    for i in movieName:
        IndianName_list.append(i.text.replace('\n',''))

    return IndianName_list[0:100]
```

```python
import pandas as pd
indian_top_movie=pd.DataFrame({})
indianName=getIndianTopMovie('https://www.imdb.com/india/top-rated-indian-movies/','td','titleColumn')
indianYear=getIndianTopMovie('https://www.imdb.com/india/top-rated-indian-movies/','span','secondaryInfo')
indianRating=getIndianTopMovie('https://www.imdb.com/india/top-rated-indian-movies/','td','ratingColumn imdbRating')
indian_top_movie['Indian Movie Name']=indianName[0:100]
indian_top_movie['Year']=indianYear[0:100]
indian_top_movie['Rating']=indianRating[0:100]

indian_top_movie
```

```python
def book_reviews(url):
    book_name=[]
    author_name=[]
    genre=[]
    book_review=[]
    bookreviews=requests.get(url)
    bookreviews_soup=BeautifulSoup(bookreviews.content)
    bookname=bookreviews_soup.find_all('h4',class_='italic')
    bookauthor=bookreviews_soup.find_all('p',class_='sans bold')
    bookgenre=bookreviews_soup.find_all('p',class_='genre-links hidden-phone')
    book_reviews=bookreviews_soup.find_all('p',class_='excerpt')
    
    for i in bookname:
        book_name.append(i.text.replace('\n',''))
    for i in bookauthor:
        author_name.append(i.text.replace('\n',''))
    for i in bookgenre:
        genre.append(i.text.replace('\n',''))
    for i in book_reviews:
        book_review.append(i.text.replace('\n',''))
    
    book={'Name':book_name,'Author':author_name,'Genre':genre,'Review':book_review}
    Book_details=pd.DataFrame(data=book)
    Book_details=Book_details[:5]
    return(Book_details)
```

```python
def Top_10_ODI_Batsmen_men(url):
    player=[]
    team=[]
    rating=[]
    odi_men=requests.get(url)
    men_10=BeautifulSoup(odi_men.content)

    players=men_10.find('div',class_='rankings-block__banner--name-large') #for rank 1
    Player_all=men_10.find_all('td',class_='table-body__cell rankings-table__name name')#for other rank
    
    teams=men_10.find_all('div',class_='rankings-block__banner--nationality')#for rank 1
    team_all=men_10.find_all('span',class_='table-body__logo-text')
    
    ratings=men_10.find_all('div',class_='rankings-block__banner--rating')#for rank 1
    rating_all=men_10.find_all('td',class_='table-body__cell rating')#for other rank

    for i in players:
        player.append(i)
    for i in Player_all:
        player.append(i.find('a').text)
  
    for i in teams:
        team.append(i.text.replace('\n','').replace(' ',''))
    for i in team_all:
        team.append(i.text.replace('\n','').replace(' ',''))

    for i in ratings:
        rating.append(i.text)
    for i in rating_all:
        rating.append(i.text)

    Top_10_men={'Player':player,'Team':team,'Rating':rating}
    Top_10_men=pd.DataFrame(data=Top_10_men)
    return(Top_10_men[:10])
```

### 6.Write a python program to scrape cricket rankings from 'icc-cricket.com'. You have to scrape:

```python

def Top_10_ODI_teams_Women(url):
    team=[]
    matches=[]
    points=[]
    ratings=[]
    
    odi10=requests.get(url)
    odi10_soup=BeautifulSoup(odi10.content)
    teamname=odi10_soup.find_all('span',class_='u-hide-phablet')
    
    matches10=odi10_soup.find_all('td',class_='rankings-block__banner--matches')
    matches10_2=odi10_soup.find_all('td',class_='table-body__cell u-center-text')
    
    points10=odi10_soup.find_all('td',class_='rankings-block__banner--points')
    points10_2=odi10_soup.find_all('td',class_='table-body__cell u-center-text')
    
    Ratings=odi10_soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
    ratings_2=odi10_soup.find_all('td',class_='table-body__cell u-text-right rating')
    
    for i in teamname:
        team.append(i.text)
        team=team[:10]
    
    for i in matches10:
        matches.append(i.text)
    
    for i in range(0,len(matches10_2),2):
        matches.append(matches10_2[i].text)
        matches=matches[:10]
    
    for i in points10:
        points.append(i.text)

    for i in range(1,len(points10_2),2):
        points.append(points10_2[i].text)
        points=points[:10]
    
    for i in Ratings:
        ratings.append(i.text.replace('\n','').replace(' ',''))

    for i in ratings_2:
        ratings.append(i.text.replace('\n','').replace(' ',''))
        ratings=ratings[:10]
    
    Top_10={'Team':team,'Matches':matches,'Points':points,'Rating':ratings}
    Top_10=pd.DataFrame(data=Top_10)
    return(Top_10)
```

#### 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def Weather_7_day_SFO(url):
    period=[]
    short=[]
    temp=[]
    desc=[]
    
    weather=requests.get(url)
    weather_soup=BeautifulSoup(weather.content)
    
    for a in weather_soup.find_all('div',class_='col-sm-2 forecast-label'):
        period.append(a.get_text(strip=True))
    for b in weather_soup.find_all(['div'], class_ = "col-sm-10 forecast-text"):
        desc.append(b.get_text(strip=True))
    shrt_desc_1 = [i.split(',')[0]for i in desc]
    shrt_desc_2 = [i.split(', ')[1]for i in desc]
    shrt_desc_2 =[i.split('.')[0]for i in shrt_desc_2]
    shrt_desc_1=[i.split('.')[-1]for i in shrt_desc_1]
    temperature = [i + '°F' for i in shrt_desc_2]
    temp = [i.split(' ',3)[-1]for i in temperature]
    
    Weather={'Period':period,'Short Description':shrt_desc_1,'Temperature':temp,'Description':desc}
    Weather=pd.DataFrame(data=Weather)
    return(Weather)
```

#### 9. Write a python program to scrape fresher job listings from internshala.com’. It should include job title, company name, CTC, and apply date.

```python
def Job_listing_fresher_internshala(url):
    page = 1
    titles =[]
    company_name=[]
    CTC=[]
    apply_date=[]
    while page != 4:
        url = f'https://internshala.com/fresher-jobs/page-{page}'
        response = requests.get(url)
        html = response.content
        soup = BeautifulSoup(html, "lxml")
        for a in soup.find_all('a',class_='link_display_like_text'):
            company_name.append(a.get_text(strip=True))
        for b in soup.find_all('div',class_='heading_4_5 profile'):
            titles.append(b.get_text(strip=True))
        for c in soup.find_all('div',class_='item_body'):
            CTC.append(c.get_text(strip=True))
        for d in soup.find_all('div',class_='item_body'):
            apply_date.append(d.get_text(strip=True))
        page = page + 1
        
    Jobs={'Job Title':titles,'Company Name':company_name,'CTC':CTC[1:313:3],'Apply by':apply_date[2:313:3]}
    Jobs=pd.DataFrame(data=Jobs)
    return(Jobs)
```

```python
Job_listing_fresher_internshala('https://internshala.com/fresher-jobs/page3')
```
